## Estimator <br>
* Machine Learning Models in scikit-learn
* Funtion : Fit and predict

```python
from sklearn.datasets import make_regression # <- Membuat DataSet Seacar Langung (Dummy)
from sklearn.linear_model import LinearRegression
```

X = Features <br>
y = Target

```python
X, y  = make_regression(n_samples=20, n_features=1, random_state=22, noise=3.0)
```

```python
import matplotlib.pyplot as plt
plt.plot (X, y, 'g.')
plt.show()
```

```python
model = LinearRegression() # <- blon di train jadi model tidak memiliki knowlage apa apa
```

## Train Model

fit() adalah fungsi pada scikit-learn yang digunakan untuk melatih model menggunakan data training. <br>

Pada proses ini, model akan:

Mempelajari hubungan antara data input dan output <br>
Mencari pola dari data yang diberikan <br>
Menyimpan hasil pembelajaran dalam bentuk parameter model <br>

```python
model.fit(X, y)
```

```python
prediksi = model.predict(X)
```

```python
plt.plot(X, y, 'b.') # <- Membuat dengan data asli
plt.plot(X, prediksi, 'r-') # <- Membuat dengan prediksi dari ai

# Menggunakan X yang sama (fitur/input tetap sama)
# y yang pertama = target asli dari data (ground truth)
# y yang kedua (prediksi) = hasil yang diperkirakan oleh model
# Jadi perbandingan ini untuk melihat seberapa bagus model mengikuti pola data

plt.show()
```

### Model dikatakan baik jika garis prediksi mampu mengikuti pola data asli, meskipun tidak semua titik tepat berada di garis karena adanya noise atau variasi data.
