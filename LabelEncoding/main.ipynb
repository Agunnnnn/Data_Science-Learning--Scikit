# DATA

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Country' : ['India', 'US', 'Japan', 'US', 'Japan'],
    'Age': [44, 34, 46, 35, 23],
    'Salary' : [72000, 65000, 98000, 45000, 43000]
})

df
```

# LABEL ENCODING

```python
from sklearn.preprocessing import  LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Country'] = le.fit_transform(df['Country'])
df
```

```python
le.classes_
```

# One Hot Encoding

```python
df = pd.DataFrame({
    'Country' : ['India', 'US', 'Japan', 'US', 'Japan'],
    'Age': [44, 34, 46, 35, 23],
    'Salary' : [72000, 65000, 98000, 45000, 34000]
})

df
```

```python
X  = df['Country'].values.reshape(-1, 1)
X
```

```python
from sklearn.preprocessing import  OneHotEncoder
```

```python
ohe = OneHotEncoder()
X = ohe.fit_transform(X).toarray()
X
```

```python
ohe.categories_
```

```python
df_ohe = pd.DataFrame(X, columns=[str(i) for i in range(X.shape[1])])
```

```python
df_ohe
```

```python
df = pd.concat([df_ohe, df],axis=1)
df
```

```python
df = df.drop(['Country'], axis=1)
df
```
