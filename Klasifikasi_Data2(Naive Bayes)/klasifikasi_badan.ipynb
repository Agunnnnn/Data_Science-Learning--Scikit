## KLASIFIKASI Menggunakan Metode Naive Bayes

## Import Library

```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split # <- spliting data / pembagian data
from sklearn.preprocessing import LabelEncoder #<- mengkonfersi data string menjadi label numrik
from sklearn.preprocessing import StandardScaler # <- mengsekala data numrik
from sklearn.naive_bayes import  GaussianNB # <- algoritma / Metode
from sklearn.metrics import confusion_matrix # <- menguji kinerja model
from sklearn.metrics import classification_report # <- evaluasi
from sklearn.metrics import  accuracy_score # <- evaluasi
```

## Baca Data

```python
df_1 = pd.read_csv("dataset_customer_churn_label_string.csv")
df_1.head() #<- menampilkan 5 data taeratas
```

```python
df_1.info() # <- mengatahui informasi terkait dataset
```

```python
df_1.empty # < - mengetahui apakah ada data kosong
```

### Process Label Encoding

ubah tipe data label yang tadinya str menjadi numrik

```python
numrik = LabelEncoder() 

df_1['churn'] = numrik.fit_transform(df_1['churn'])

df_1['city'] = numrik.fit_transform(df_1['city'])

df_1['plan_type'] = numrik.fit_transform(df_1['plan_type'])
df_1.head()
```

### Data sudah berubah dimana yagn tadinya string sekarang int

## Menetukan X dan Y

```python
x = df_1.iloc[:, :-1].values
y = df_1.iloc[:, -1].values
```

```python
x
```

```python
y
```

### Spiliting Data


```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=123)

print("X Train = ", len(x_train))
print("X Test = ", len(x_test))
print("Y Train = ", len(y_train))
print("Y Test = ", len(y_test))
```

### Pengskalaan Nilai

```python
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

## Clasifikasi Model

```python
klasifikasi = GaussianNB()
```

```python
klasifikasi.fit(x_train, y_train)
```

```python
y_predict = klasifikasi.predict(x_test)
y_predict
```

```python
klasifikasi.predict_proba(x_test)
```

## EVALUASI

```python
cm = confusion_matrix(y_test,y_predict)
print(cm)
```

```python
akurasi = classification_report(y_test, y_predict)
print(akurasi)
```

```python
akurasi = accuracy_score(y_test, y_predict)
print("Tingkat Akurasi : %d Persen"%(akurasi*100))
```
