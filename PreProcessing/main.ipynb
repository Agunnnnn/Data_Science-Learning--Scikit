```python
import numpy as np
from sklearn import preprocessing
```

# SAMPLE DATA

```python
data = np.array([[2.1, -1.9, 5.5],
                [-1.5, 2.4, 3.5],
                [0.5 , -7.9, 5.6],
                [5.9 , 2.3 ,-5.8]])

data
```

```python
data.shape
```

# 1 Binarisation

```python
preprocessor = preprocessing.Binarizer(threshold=0.5)
data_binarisation = preprocessor.transform(data)

data_binarisation
```

# 2. Scalling

```python
preprocessor= preprocessing.MinMaxScaler(feature_range=(0,1))
preprocessor.fit(data)
scalling_data = preprocessor.transform(data)
scalling_data
```
